# Top 10 highest rated iPhones on Flipkart in india?

```python
highest_rated = dataset.sort_values(by = ["Star Rating"], ascending = False)
highest_rated = highest_rated.head(10)
```

```python
highest_rated["Product Name"]
```

# How many ratings do the highest rated iPhones on Flipkart have?

```python
plt.figure(figsize = (12,6))
sns.barplot(x=highest_rated["Product Name"], y=highest_rated["Number Of Ratings"])
plt.xticks(rotation=-80)
plt.title("number of rating of highest rated iphones")
plt.show()
```

# Which iphone has the highest number of reviews on Flipkart?

```python
dataset.head(5)
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x=highest_rated["Product Name"], y=highest_rated["Number Of Reviews"], color="g")
plt.xticks(rotation=-80)
plt.title("number of reviews of highest rated iphones")
plt.show()
```

# What is the relationship between the sale price of iphone and the number of ratings on Flipkart?

```python
dataset.head(5)
```

```python
sns.scatterplot(data=dataset, x="Number Of Ratings", y="Sale Price", size="Discount Percentage", color ="r")
plt.title("Relationship between Sale price and Number of Ratings of iphones")
plt.show()
```

There is negative relationship between the sale price of iphone and number of ratings. It mean lower sale prices are sold more in India.

# What is the relationship between the discount percentage and the number of ratings of iphones on Flipkart?

```python
sns.scatterplot(data=dataset, x="Number Of Ratings", y="Discount Percentage", size="Sale Price")
plt.title("Relationship between Discount and Number of Ratings of iphones")
plt.show()
```

# least expensive and most expensive 

```python
most_expensive = dataset.loc[dataset["Sale Price"].idxmax()]
least_expensive = dataset.loc[dataset["Sale Price"].idxmin()]

# Display the result

print("Most Espensive Product:")
print(most_expensive)
print("Least Espensive Product:")
print(least_expensive)
```
